### 1. Idade media das mães e dos pais por município (coluna munResNome)


```python
sinasc_raw['IDADEPAI'].count()
```

```python
sinasc_raw['IDADEMAE'].count()
```

### 2. Peso médio dos bebes por sexo que nasceram no dia do seu aniversário por faixas de escolaridade mae
Ex: Você, aluna(o), nasceu no dia 10/01, então você precisa filtrar o conjunto de dados nessa data e calcular o peso médio dos bebês de cada sexo por faixa de escolaridade da mãe.

```python
sinasc_2 = sinasc_raw[['DTNASC', 'PESO', 'SEXO', 'ESCMAE']]
sinasc_2 = pd.DataFrame(sinasc_2)
sinasc_2
```

```python
sinasc_2.loc[:,'DTNASC'] = pd.to_datetime(sinasc_2['DTNASC'])
```

```python
sinasc_2['DTNASC'] = sinasc_2['DTNASC'].dt.strftime('%m-%d')
```

```python
sinasc_2509 = sinasc_2[sinasc_2['DTNASC'] == "09-25"]
```

```python
sinasc_pesomedio_2509 = sinasc_2509.groupby(['ESCMAE', 'SEXO'])['PESO'].mean()
sinasc_pesomedio_2509 = pd.DataFrame(sinasc_pesomedio_2509)
sinasc_pesomedio_2509
```

### 3. Qual o municipio que nasceu menos bebe em 2019?
    - qual a idade media, maxima, minima das maes nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?

```python
sinasc_3 = sinasc_raw[['DTNASC', 'IDADEPAI', 'IDADEMAE', 'munResNome']]
sinasc_3 = pd.DataFrame(sinasc_3)
```

```python
sinasc_3['IDADEMAE'].count()
```

```python
sinasc_3.groupby('munResNome')['IDADEMAE'].count().sort_values()
```

```python
sinasc_castanheiras = sinasc_3[sinasc_3['munResNome'] == "Castanheiras"]
```

```python
sinasc_castanheiras.dropna().groupby('munResNome').agg(['min', 'max', 'mean'])
```

### 4. Qual o municipio que nasceu mais bebe no mês de março?
    - qual a quantidade de filhos vivos media, maxima, minima nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?



```python
sinasc_4 = sinasc_raw[['DTNASC', 'IDADEPAI', 'munResNome', 'QTDFILVIVO']]
```

```python
sinasc_4.loc[:,'DTNASC'] = pd.to_datetime(sinasc_4['DTNASC'])
```

```python
sinasc_4['DTNASC'] = sinasc_4['DTNASC'].dt.strftime('%m')
```

```python
sinasc_marco = sinasc_4[sinasc_4['DTNASC'] == "03"]
```

```python
sinasc_marco.groupby('munResNome')['DTNASC'].count().sort_values(ascending=False)
```

```python
sinasc_marco_portovelho = sinasc_marco[sinasc_marco['munResNome'] == "Porto Velho"]
```

```python
sinasc_marco_portovelho
```

```python
sinasc_marco_portovelho['QTDFILVIVO'].count()
```

```python
sinasc_marco_portovelho = sinasc_marco_portovelho.dropna(subset=['QTDFILVIVO'])
sinasc_marco_portovelho
```

```python
sinasc_marco_portovelho.groupby(sinasc_marco['munResNome']).agg({'QTDFILVIVO':['min','max','mean']})
```

```python
sinasc_marco_portovelho = sinasc_marco_portovelho.dropna(subset=['IDADEPAI'])
```

```python
sinasc_marco_portovelho.groupby(sinasc_marco['munResNome']).agg({'IDADEPAI':['min','max','mean']})
```
